```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
data = pd.read_csv("D:\Data Science\Task MileStone 1\Coffee.csv")
data.head()
```

```python
# Calculate the total percentage of null values for each column
total_null_percentage = data.isnull().sum() / len(data) * 100

# Display the results
print("Total percentage of null values for each column:")
for column, percentage in total_null_percentage.items():
    print(f"{column}: {percentage:.2f}%")
```

```python
# Columns to be removed based on relevance and missing values
columns_to_remove = [
    'Lot.Number', 'Mill', 'ICO.Number', 'Owner.1', 'Unnamed: 0',
    'Certification.Address', 'Certification.Contact', 
    'Certification.Body', 'Expiration', 'In.Country.Partner'
]
df_cleaned = data.drop(columns=columns_to_remove)

# Verify columns are removed
print("Remaining Columns:\n", df_cleaned.columns)
df_cleaned
```

```python
numerical_cols = df_cleaned.select_dtypes(include=['number']).columns

# Function to detect outliers using IQR method
def detect_outliers_iqr(df_cleaned, column):
    Q1 = df_cleaned[column].quantile(0.25)
    Q3 = df_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df_cleaned[(df_cleaned[column] < lower_bound) | (df_cleaned[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Detecting outliers for each numerical column
outliers_dict = {}
bounds_dict = {}
for col in numerical_cols:
    outliers, lower_bound, upper_bound = detect_outliers_iqr(df_cleaned, col)
    outliers_dict[col] = outliers
    bounds_dict[col] = (lower_bound, upper_bound)

# Displaying the number of outliers detected in each column
for col, outliers in outliers_dict.items():
    print(f'Number of outliers in {col}: {len(outliers)}')
```

```python
# Replace null values
for col in numerical_cols:
    df_cleaned[col].fillna(df_cleaned[col].median(), inplace=True)  # Impute null values with median

# Function to remove outliers using IQR method
def remove_outliers_iqr(df_cleaned, column):
    Q1 = df_cleaned[column].quantile(0.25)
    Q3 = df_cleaned[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) & (df_cleaned[column] <= upper_bound)]
    return df_cleaned

# Remove outliers for each numerical column
for col in numerical_cols:
    df_cleaned = remove_outliers_iqr(df_cleaned, col)

# Displaying the shape of the DataFrame after removing outliers
print("Shape of DataFrame after removing outliers:", df_cleaned.shape)
```

```python
categorical_cols = df_cleaned.select_dtypes(include=['object']).columns.tolist()
# Iterate through each categorical column
for col in categorical_cols:
    # Check for unique values to identify inconsistencies
    unique_values = data[col].unique()
    print(f"Unique values in {col}: {unique_values}")

    # Rectify inconsistencies
    # Example: Convert text to title case and remove leading/trailing whitespaces
    data[col] = data[col].str.strip().str.title()

    # Handle missing values if necessary
    # Example: Impute missing values with the mode
    mode_value = data[col].mode()[0]
    data[col].fillna(mode_value, inplace=True)

    # Confirm corrections
    corrected_values = data[col].unique()
    print(f"After correction, unique values in {col}: {corrected_values}")
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_cleaned.select_dtypes(include=['number']), orient="v")
plt.title('Box plot of Numerical Columns')
plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees
plt.tight_layout()
plt.show()
```

```python
# Display the results
print("Number of unique values for each column:")
for column in df_cleaned.columns:
    unique_values_count = df_cleaned[column].nunique()
    print(f"{column}: {unique_values_count}")

print("\nPercentage of unique values for each column:")
for column in df_cleaned.columns:
    unique_values_count = df_cleaned[column].nunique()
    unique_percentage = (unique_values_count / len(df_cleaned) * 100)
    print(f"{column}: {unique_percentage:.2f}%")
```

```python
# Replacement by mean for Numerical Data USING Median and Non-Numerical Data USING Mode :
for column in df_cleaned.columns:
    if df_cleaned[column].dtype == 'object':
        mode_value = df_cleaned[column].mode()[0]
        print(f"Filling missing values in column '{column}' with mode: {mode_value}")
        df_cleaned[column].fillna(mode_value, inplace=True)
    else:
        median_value = df_cleaned[column].mean()
        print(f"Filling missing values in column '{column}' with mean: {median_value}")
        df_cleaned[column].fillna(median_value, inplace=True)
```

```python
# Standardize all categorical columns
for column in df_cleaned.select_dtypes(include=['object']).columns:
    df_cleaned[column] = df_cleaned[column].str.strip().str.title()

# Verify changes
df_cleaned
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA


selected_columns = [
    'Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 
    'altitude_low_meters', 'altitude_high_meters', 'altitude_mean_meters', 'Total.Cup.Points'
]

# Impute missing values
imputer = SimpleImputer(strategy='mean')
imputed_data = imputer.fit_transform(df_cleaned[selected_columns])

# Standardize the data
x = StandardScaler().fit_transform(imputed_data)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(x)

# Create DataFrame for PCA results
principal_df = pd.DataFrame(data=principal_components, columns=['principal component 1', 'principal component 2'])

# Combine PCA results with 'Total.Cup.Points'
final_df = pd.concat([principal_df, df_cleaned[['Total.Cup.Points']]], axis=1)

# Plot PCA
plt.figure(figsize=(8, 8))
plt.scatter(final_df['principal component 1'], final_df['principal component 2'], c=final_df['Total.Cup.Points'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 component PCA')
plt.colorbar(label='Total Cup Points')
plt.show()
```

```python
from scipy.stats import chi2_contingency

# Example categorical columns (replace with your actual categorical columns)
categorical_columns = ['Processing.Method', 'Color', 'Species', 'Owner', 'Country.of.Origin', 
                       'Farm.Name',  'Company', 'Region']

# Create a list to store chi-square test results
chi2_results = []

# Perform chi-square test for each pair of categorical variables
for col1 in categorical_columns:
    for col2 in categorical_columns:
        if col1 != col2:
            # Create contingency table
            contingency_table = pd.crosstab(df_cleaned[col1], df_cleaned[col2])
            # Perform chi-square test
            chi2, p, dof, ex = chi2_contingency(contingency_table)
            # Store results
            chi2_results.append((col1, col2, chi2, p))

# Convert results to DataFrame
chi2_df = pd.DataFrame(chi2_results, columns=['Column 1', 'Column 2', 'Chi-square', 'p-value'])

# Display results
print(chi2_df)
```

```python
from itertools import combinations
from scipy.stats import ttest_ind

# Identify pairs of columns containing continuous variables
continuous_columns = ['Aroma', 'Flavor', 'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 
                      'Clean.Cup', 'Sweetness', 'Cupper.Points', 'Total.Cup.Points', 'Moisture', 
                      'Category.One.Defects', 'Quakers', 'Category.Two.Defects', 'altitude_low_meters',
                      'altitude_high_meters', 'altitude_mean_meters']  # Add more as needed

# Perform t-test for each pair of columns
column_pairs = combinations(continuous_columns, 2)
for pair in column_pairs:
    t_statistic, p_value = ttest_ind(df_cleaned[pair[0]], df_cleaned[pair[1]])
    print(f"t-test for {pair}: t-statistic = {t_statistic}")
```
